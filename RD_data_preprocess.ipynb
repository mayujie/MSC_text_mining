```
def find_noise(reg, datalist):
  count = 0
  found_list = []

  for i, row in enumerate(datalist):
    found = re.findall(reg, row)

    if found != []:
      count += 1
      # print(found[0])
      # found = list(filter(None, found[0]))
      # print(found)
      found_list.append([i ,found])
  
  return count, found_list
  # return found_list
```

### categorize the URL (6 classes)

1.  **|facebook| --> \<FACEBOOKURL\>,**  
2.  **|book| --> \<BOOKURL\>,**
3.  **|youtube  youtu  youtu.be| --> \<YOUTUBEURL\>,**
4.  **|google  goo\.gl  goo\.| --> \<GOOGLEURL\>,**
5.  **|amzn  amazon| --> \<AMAZONURL\>,**
6.  **|rest of URL| --> \<URL\>**

```
## find keywords in found_url
def found_key_url(reg, datalist):
  found_key = []
  for i in datalist:
    for el in i[1]:
      # print(re.findall(r'amazon', el))
      # if re.findall(r'amazon|amzn', el): # |amzn  amazon|  |facebook|  |book|   |youtube  youtu  youtu.be|   |google  goo\.gl  goo\.|
      if re.findall(reg, el):
        found_key.append(i)
        # print(i)
  return found_key
```

```
def find_noise(reg, datalist):
  count = 0
  found_list = []

  for i, row in enumerate(datalist):
    found = re.findall(reg, row)

    if found != []:
      count += 1
      # print(found[0])
      # found = list(filter(None, found[0]))
      # print(found)
      found_list.append([i ,found])
  
  return count, found_list
  # return found_list
```
